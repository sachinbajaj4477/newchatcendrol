```python
# Import necessary libraries
import pandas as pd
from langchain_groq import ChatGroq
from langchain_experimental.agents.agent_toolkits import create_csv_agent

# Define API key and initialize the ChatGroq model
groq_api = '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'
llm = ChatGroq(
    temperature=0.3,           # Balanced temperature for deterministic yet flexible responses
    model="llama3-70b-8192",   # High-performance LLM for accurate outputs
    api_key=groq_api,
    max_tokens=1000,           # Limit response length
    top_p=0.9                  # Focus on the most relevant results
)

# Specify your CSV file path
csv_file_path = 'data.csv'  # Update this with your CSV file's path

# Create the CSV agent without a custom prompt
agent = create_csv_agent(
    llm=llm,
    path=csv_file_path,
    verbose=True,               # Enables detailed logs for debugging
    max_iterations=15,          # Limit the number of reasoning steps
    max_execution_time=30,      # Set a timeout of 30 seconds for responses
    allow_dangerous_code=True,  # Allow executing Python code for advanced queries
    return_intermediate_steps=False  # Do not show intermediate reasoning steps
)

# Function to query the CSV data
def query_data(query):
    try:
        # Send the query to the CSV agent
        response = agent.invoke(query)
        return response
    except Exception as e:
        return f"Error: {str(e)}"

# Start a simple CLI loop for live chat
def live_chat():
    print("Welcome to the live chat! Type your query below (type 'exit' to quit):")
    while True:
        # Get user input
        query = input("You: ")
        if query.lower() == "exit":
            print("Exiting chat. Goodbye!")
            break

        # Get the response from the model
        response = query_data(query)

        # Print the response
        print(f"AI: {response}")

# Run the live chat
if __name__ == "__main__":
    live_chat()
```
